```python
import gradio as gr
from transformers import pipeline
```

```python
sentiment = pipeline("sentiment-analysis")
```

```python
def findSentiment(input):
  return sentiment(input)
```

```python
result = findSentiment("You are jokingly good!")
```

```python
print(result)
```

```python
iface = gr.Interface(fn=findSentiment,
                     inputs="text",
                     outputs=["text"],title="MY Sentiment Analyzer",
                     theme="darkhuggingface"
                     )
```

```python
iface.launch()
```
